```python
import pandas as pd
import numpy as np
import pandas_profiling
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LassoCV
from sklearn import datasets
from collections import Counter


#Import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB

#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score
```

# Escenarios
De acuerdo al problema inicial la clasificacion se realizaria en las categorias: Agua, bosque, Agua o bosque, ninguno. Al ejecutar las tecnicas de clasificación logramos observar que el problema estaba siendo facilmente calculado, razon por la cual se decidio añadir un grado de complejidad al asunto añadiendo una clasificación adicional: cloud.
A continuacion vamos a cargar los 2 datasets, procesarlos y hacer una comparación entre sus estadisticas.

```python
# satellite = pd.read_csv('satellite_dataset.csv')
satellite_nubes = pd.read_csv('satellite_dataset-con-nubes.csv')
satellite_nubes.head()
```

```python

# Reemplazar el valor de columna Wofs por NaN(0) y Wofs(1) dataset satellite_nubes
satellite_nubes['wofs'] = satellite_nubes['wofs'].replace(0,np.nan)
satellite_nubes['wofs'] = satellite_nubes['wofs'].replace(1,'wofs')
```

```python

# Reemplazar el valor de columna bosque por NaN(0) y bosque(1) dataset satellite_nubes
satellite_nubes['bosque'] = satellite_nubes['bosque'].replace(0,np.nan)
satellite_nubes['bosque'] = satellite_nubes['bosque'].replace(1,'bosque')
```

```python

# Reemplazar el valor de columna wofs_bosque por NaN(0) y wofs_bosque(1) dataset satellite_nubes
satellite_nubes['wofs_bosque'] = satellite_nubes['wofs_bosque'].replace(0,np.nan)
satellite_nubes['wofs_bosque'] = satellite_nubes['wofs_bosque'].replace(1,'wofs_bosque')
```

```python

# Reemplazar el valor de columna ninguno por NaN(0) y ninguno(1) dataset satellite_nubes
satellite_nubes['ninguno'] = satellite_nubes['ninguno'].replace(0,np.nan)
satellite_nubes['ninguno'] = satellite_nubes['ninguno'].replace(1,'ninguno')
```

```python
# Reemplazar el valor de columna cloud por NaN(0) y cloud(1) dataset satellite_nubes
satellite_nubes['cloud'] = satellite_nubes['cloud'].replace(0,np.nan)
satellite_nubes['cloud'] = satellite_nubes['cloud'].replace(1,'cloud')
```

```python
satellite_nubes.head(10)
```

```python
#Se crea una nueva columna para almacenar el valor de etiqueta

satellite_nubes['label']=np.nan
```

```python
#Se indica en la columna 'label', la clasificacion de la columna 'wofs'. Se esribe "Wofs" en columna 'label' si en  
#la columna 'wofs' esta este valor
for label in [col for col in satellite_nubes.columns if 'wofs' in col]:
    satellite_nubes['label'].fillna(satellite_nubes[label],inplace=True)
    
#Se indica en la columna 'label', la clasificacion de la columna 'bosque'. Se esribe "bosque" en columna 'label' si en  
#la columna 'bosque' esta este valor
for label in [col for col in satellite_nubes.columns if 'bosque' in col]:
    satellite_nubes['label'].fillna(satellite_nubes[label],inplace=True)
    
#Se indica en la columna 'label', la clasificacion de la columna 'wofs_bosque'. Se esribe "wofs_bosque" en columna 'label' si en  
#la columna 'wofs_bosque' esta este valor
for label in [col for col in satellite_nubes.columns if 'wofs_bosque' in col]:
    satellite_nubes['label'].fillna(satellite_nubes[label],inplace=True)
    
#Se indica en la columna 'label', la clasificacion de la columna 'ninguno'. Se esribe "ninguno" en columna 'label' si en  
#la columna 'ninguno' esta este valor
for label in [col for col in satellite_nubes.columns if 'ninguno' in col]:
    satellite_nubes['label'].fillna(satellite_nubes[label],inplace=True)
    
#Se indica en la columna 'label', la clasificacion de la columna 'cloud'. Se esribe "cloud" en columna 'label' si en  
#la columna 'cloud' esta este valor
for label in [col for col in satellite_nubes.columns if 'cloud' in col]:
    satellite_nubes['label'].fillna(satellite_nubes[label],inplace=True)
satellite_nubes.head(20)
```

```python
# Se crea el conjunto de calsificacion y el de entrenamiento
Y = satellite_nubes["label"]
X = satellite_nubes.drop(columns=["label","wofs","bosque","wofs_bosque","ninguno","cloud"], axis=1)
```

# Se crean los conjuntos de entrenamiento y de prueba
Para hacer esto se deben dividir los datos en 4 conjuntos de datos como se puede apreciar en la siguiente imagen.
![title](conjuntos.png)

```python
# Se dividen los datos en los conjuntos de Train y test
X_train, X_test, y_train, y_test = train_test_split(
    X,
    Y,
    test_size=0.3,
    random_state=0
)
```

# Naive Bayes para multiples capas
La clasificación multicapa del algoritmo Naive Bayes es conocida como "Clasificación Naive Bayes Multinomial". Como se puede apreciar la clasificación de nuestro conjunto de caracteristicas puede tener 4 valores: wofs, bosque, wofs_bosqu y ninguno.

# Calculo de estadisticas con dataset satellite_nubes

```python
#Create a Gaussian Classifier
gnb_nubes = GaussianNB()
#gnb_nubes = MultinomialNB()
```

```python
#Train the model using the training sets
gnb_nubes.fit(X_train, y_train)
```

```python
#Predict the response for test dataset
y_pred = gnb_nubes.predict(X_test)
```

```python
# Model Accuracy
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
#Counter(y_pred)
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
